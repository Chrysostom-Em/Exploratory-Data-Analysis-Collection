```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
nyc_school = pd.read_csv('schools.csv')
nyc_school.head()
```

1. Subset the data to find the schools with math scores of at least 80%, considering the maximum possible score is 800 and save this as a pandas DataFrame called best_math_schools.


```python
math_scores = nyc_school[['school_name','average_math']].reset_index()
math_scores.head()
```

```python
best_math_schools = math_scores[math_scores['average_math']>= 80 / 100 * 800]
best_math_schools.sort_values(by='average_math', ascending=False)
```

2. Find the 10 best performing schools based on the total score across the three SAT sections.

```python
nyc_school['total_SAT'] = nyc_school['average_math'] + nyc_school['average_reading'] + nyc_school['average_writing']

top_schools = nyc_school[['school_name', 'total_SAT']].sort_values(by='total_SAT', ascending=False).reset_index()
top_schools.head(10)
```

3. Find out the number of schools, average SAT, and standard deviation of SAT for the NYC borough with the largest standard deviation, rounded to two decimal places

```python
nyc_borough = nyc_school.groupby('borough')['total_SAT'].agg(['count', 'mean', 'std']).round(2)
nyc_borough
```
